Write a python program to print the contents of a directory using the os modules

```python
import os

def list_directory_contents(directory_path):
    try:
        # Get the list of files and directories
        contents = os.listdir(directory_path)

        # Print the contents
        print(f"Contents of directory '{directory_path}':")
        for item in contents:
            print(item)
    except FileNotFoundError:
        print(f"The directory '{directory_path}' does not exist.")
    except PermissionError:
        print(f"Permission denied to access the directory '{directory_path}'.")

# Replace 'your_directory_path' with the path of the directory you want to list
your_directory_path = '/'
list_directory_contents(your_directory_path)
```

## Variable and Datatypes



**Variable:** A variable in programming is like a labeled container that holds data. You can store different types of information (like numbers, text, or truth values) in variables, and change their contents as needed throughout your program.

**Variable Naming Rules:**

---



Variable names can contain letters, numbers, and underscores.
They cannot start with a number.
Python variables are case-sensitive (age and Age are different).


**Types:** Python is dynamically typed, meaning you don't need to declare the type of a variable when you define it. Python infers the type based on the value assigned. bold text

```python
# Examples
name = "Alice"  # String
age = 30  # Integer
is_student = False  # Boolean
```

**Type Conversion:** You can convert variables from one type to another using type casting functions like int(), float(), str(), etc.

```python
# Example
x = "5"
y = int(x)  # Converts string '5' to integer 5
```

**Printing Variables:** To print the value of a variable, you can use the print() function.

```python
# Example
age = 30
print(age)  # Output: 30
```

**Variable Reassignment:** You can change the value of a variable simply by assigning it a new value.

```python
# Example
x = 5
x = x + 1
print(x)  # Output: 6
```

**Constants:** Python doesn't have built-in constant types, but conventionally, constants are defined in uppercase to indicate they shouldn't be changed.

```python
PI=3.14
```

**String Slicing **" String slicing in Python is a technique used to extract a substring (a portion of a string) by specifying indices within square brackets []."

```python
# Define a string
s = "Hello, World!"

# Basic slicing: s[start:end]
substring = s[3:12]  # " o,World"

# Omitting start or end defaults to beginning or end
substring_start = s[:5]   # "Hello"
substring_end = s[7:]     # "World!"

# Negative indices count from end of string
substring_negative = s[-6:-1]  # "World"

# Slicing with a step (third parameter)
substring_step = s[::2]   # "Hlo ol!"

# Reverse a string
reverse_string = s[::-1]  # "!dlroW ,olleH"

print(s)
print(substring)
print(substring_start)
print(substring_end)
print(substring_negative)
print(substring_step)
print(reverse_string)
```

**Lists and Tuples in Python**

Lists and tuples are both used to store collections of items in Python. The main difference between them is that lists are mutable, meaning their elements can be changed, while tuples are immutable, meaning once they are created, their elements cannot be changed.
